```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading the data
data = pd.read_csv('LassovsRidge.csv') 
X = data[['X1','X2','X3','X4','X5']].values
t = data[['target']].values
```

```python
# Initializing the models
ridge_reg = Ridge()
lasso_reg = Lasso()

# Training both Ridge and Lasso for different alpha values and comparing them
for alpha in range(11):
    ridge_reg.alpha = alpha
    ridge_reg.fit(X,t)
    lasso_reg.alpha = alpha
    lasso_reg.fit(X,t)
    plt.plot(range(1, len(ridge_reg.coef_) + 1), ridge_reg.coef_, color = "red")
    plt.plot(range(1, len(lasso_reg.coef_) + 1), lasso_reg.coef_, color = "blue")

plt.title('Coefficients')
plt.show()
```
